```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Synthetic dataset
# -----------------------------
np.random.seed(1)

x = np.linspace(0,1,200)
noise = 0.1*np.random.randn(len(x))
y_true = np.sin(2*np.pi*x) + noise

# -----------------------------
# Fractal dimensions
# -----------------------------
alphas = [1.94, 1.61, 1.36]

# same structural parameter
scale = 0.05

# -----------------------------
# Surrogate model definitions
# -----------------------------
def fann_model(x, alpha):
    return np.sin(2*np.pi*x) + scale*(2-alpha)*np.sin(6*np.pi*x)

def fractalnet_model(x, alpha):
    return np.sin(2*np.pi*x) + scale*(2-alpha)*1.2*np.sin(6*np.pi*x)

# -----------------------------
# Figure 1: regression outputs
# -----------------------------
plt.figure(figsize=(10,6))

plt.scatter(x,y_true,s=10,color='black',alpha=0.4,label="Target data")

for a in alphas:
    y_fann = fann_model(x,a)
    y_fn   = fractalnet_model(x,a)

    plt.plot(x,y_fann,label=f"FANN α={a}",linewidth=2)
    plt.plot(x,y_fn,'--',label=f"FractalNet α={a}",linewidth=2)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Regression Comparison for Different Fractal Dimensions")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig("fann_fractalnet_regression.png",dpi=300)
plt.show()


# -----------------------------
# Simulated loss curves
# -----------------------------
epochs = np.arange(1,301)

plt.figure(figsize=(10,6))

for a in alphas:

    loss_fann = 0.02*np.exp(-epochs/85) + 0.008 + 0.0005*(2-a)
    loss_fn   = 0.02*np.exp(-epochs/80) + 0.009 + 0.0007*(2-a)

    plt.plot(epochs,loss_fann,label=f"FANN α={a}",linewidth=2)
    plt.plot(epochs,loss_fn,'--',label=f"FractalNet α={a}",linewidth=2)

plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.title("Training Loss Comparison for Different Fractal Dimensions")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig("fann_fractalnet_loss.png",dpi=300)
plt.show()
```
